### Filtering data

```python
filt = df["Region"] == "West"
df[filt]
```

```python
filt = (df["Region"] == "West") & (df["Zip"] > 90000)

df.loc[filt, ["First Name", "Last Name"]]
```

### Manipulating and making changes to data

### Changing column headers

```python
df.rename(columns={"E Mail": "EMail Address"}, inplace=True) 
```

```python
df.columns = [x.upper() for x in df.columns]
df
```

### Changing data

```python
df["PASSWORD"] = "default"
df
```

```python
filt = df["FIRST_NAME"] == "Stephen"
df.loc[filt, "LAST_NAME"] = "Hawking"
```

### Adding Columns

```python
df["Full name"] = df["FIRST_NAME"] + " " + df["LAST_NAME"]
df
```

### Removing columns

```python
#df.drop("New col", axis="columns")
df.drop(859327)
```

### Sorting data

```python
df.sort_values(by="FIRST_NAME", ascending=False)
```

### Grouping data

```python
group = df.groupby("REGION")
```
